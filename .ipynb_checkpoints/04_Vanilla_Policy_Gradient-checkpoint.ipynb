# Vanilla Policy Gradient

## Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import time
from IPython.display import display, clear_output

import gymnasium as gym

from collections import defaultdict

import torch
import torch.nn as nn
import torch.nn.functional as F
```

## Config

```python
CONFIG = {
    'VPG': {
        'linear': {
            'gamma': 0.99,
            'num_episodes': 5000,
            'policy_lr': 1e-3,
            'val_fn_lr': 1e-3,
            'val_fn_epochs': 10,
        },
        'mlp': {    
            'hidden_sizes': [32],
            'gamma': 0.99,
            'num_episodes': 1000,
            'policy_lr': 1e-3,
            'val_fn_lr': 1e-3,
            'val_fn_epochs': 10,
        }
    },
    'animation': {
        'FPS': None,
        'do_truncate': True,
    },
    'plot_stats_lin_size': 100,
    'plot_stats_mlp_size': 20,
}
```

## Creating the environment

```python
env = gym.make('CartPole-v1', render_mode = 'rgb_array')
```

## Helper functions

```python
def animate_policy(env, policy: np.ndarray, FPS: int = 12, do_truncate: bool = True):
    '''
    Animates using the builtin env.render() function

    Args:
        env: OpenAI Env
        policy (np.ndarray): Array of size [S x A] representing the policy
        FPS (int): The frames per second of the animation
        do_truncate (bool): If True, the animation will be stopped after 100 frames
    '''

    def interpret_state(s):
        return f"({s[0]:.3f}m, {s[2]*180/np.pi:.1f}°) "

    figure_size = (6,4)
    
    state_dims = env.observation_space.shape[0]
    act_dims = env.action_space.n
    
    s, _ = env.reset()
    actions = {0:"←", 1:"→"}
    step = 0
    while True:
        
        action = np.random.choice(list(range(act_dims)), p = policy(torch.as_tensor(s)).detach().numpy())
        step += 1
        clear_output(wait = True)
        plt.figure(figsize = figure_size)
        plt.imshow(env.render())
        plt.title(f'State: {interpret_state(s)}, Action: {actions[action]}, Time: {step}s')
        plt.axis('off')
        plt.show()

        if FPS is not None:
            time.sleep(1/FPS)
        
        s, r, terminated, truncated, _ = env.step(action)
        if terminated or (truncated and do_truncate):
            break
    
    # Show final frame
    clear_output(wait = True)
    plt.figure(figsize= figure_size)
    plt.imshow(env.render())
    plt.title(f'Final State: {interpret_state(s)}, Total Time: {step}s')
    plt.axis('off')
    plt.show()
```

```python
def plot_stats(stats, window_size = 25):
    def moving_average(x, window_size=25):
        return np.convolve(x, np.ones(window_size)/window_size, mode='valid')

    fig, (ax1, ax2) = plt.subplots(1,2, figsize = (12,4))
    
    episode_lengths = stats['ep_length']
    smoothed_lengths = moving_average(episode_lengths, window_size=window_size)
    
    ax1.plot(smoothed_lengths)
    ax1.set_title("Smoothed Episode Lengths")
    ax1.set_xlabel("Episode")
    ax1.set_ylabel("Length")

    val_loss = stats['val_loss']
    smoothed_loss = moving_average(val_loss, window_size=window_size)
    
    ax2.plot(smoothed_loss)
    ax2.set_title("Smoothed Value Function Loss")
    ax2.set_xlabel("Episode")
    ax2.set_ylabel("Loss")
    
    plt.show()
    
    print(f"{stats['completions']} Completions")
```

## Defining Policy Functions

```python
class LinearPolicy(nn.Module):
    def __init__(self, state_dims, act_dims):
        super().__init__()
        self.linear = nn.Linear(state_dims, act_dims)

    def forward(self, state):
        logits = self.linear(state)
        return F.softmax(logits, dim = -1)

class MLPPolicy(nn.Module):
    def __init__(self, state_dims, act_dims, hidden_sizes=[32]):
        super().__init__()
        layer_sizes = [state_dims] + hidden_sizes + [act_dims]
        layers = []
        activation = nn.ReLU 
        output_activation = nn.Identity 

        for i in range(len(layer_sizes) - 1):
            layers.append(nn.Linear(layer_sizes[i], layer_sizes[i+1])) 
            if i < len(layer_sizes) - 1: # Add activation for all but the very last layer
                if i < len(layer_sizes) - 2: # For hidden layers
                    layers.append(activation())
                else: # For the output layer
                    layers.append(output_activation())

        self.mlp = nn.Sequential(*layers)

    def forward(self, state):
        logits = self.mlp(state)
        return F.softmax(logits, dim = -1)
```

## Defining the Value Function

```python
class MLPValue(nn.Module):
    def __init__(self, state_dims, hidden_sizes=[16]):
        super().__init__()
        layer_sizes = [state_dims] + hidden_sizes + [1]
        layers = []
        activation = nn.ReLU 
        output_activation = nn.Identity 

        for i in range(len(layer_sizes) - 1):
            layers.append(nn.Linear(layer_sizes[i], layer_sizes[i+1])) 
            if i < len(layer_sizes) - 1: # Add activation for all but the very last layer
                if i < len(layer_sizes) - 2: # For hidden layers
                    layers.append(activation())
                else: # For the output layer
                    layers.append(output_activation())

        self.mlp = nn.Sequential(*layers)

    def forward(self, state):
        return self.mlp(state)
```

## Vanilla Policy Gradient

```python
def VanillaPolicyGradient(env, policy, gamma = 1.0, num_episodes = 10, policy_lr = 1e-4,
              val_fn_lr = 1e-3, val_fn_epochs = 10):

    s_dims = env.observation_space.shape[0]
    a_dims = env.action_space.n
    
    stats = {'ep_length': [],
             'completions': 0,
             'val_loss': []}

    
    value_fn = MLPValue(s_dims)
    mse_loss = nn.MSELoss()
    
    policy_optimizer = torch.optim.Adam(policy.parameters(), lr = policy_lr)
    value_optimizer = torch.optim.Adam(value_fn.parameters(), lr = val_fn_lr)
    
    for e in range(1, num_episodes + 1):

        if e % 1 == 0 or e == num_episodes:
            print(f'\rEpisode : {e}/{num_episodes} ({100*e/num_episodes:.2f}%)', end = '')
            sys.stdout.flush()

        H = defaultdict(lambda: [])
        T = 0
        s, _ = env.reset()
        while True:
            T += 1
            probs = policy(torch.as_tensor(s))
            a = np.random.choice(list(range(a_dims)), p = probs.detach().numpy())

            s_n, r, terminated, truncated, _ = env.step(a)

            H['s'] += [s]
            H['a'] += [a]
            H['r'] += [r]
            H['log_probs'] += [torch.log(probs[a])]
            
            if terminated or truncated:
                stats['ep_length'] += [T]
                if truncated:
                    stats['completions'] += 1
                break

            s = s_n

        R = 0.0
        H['R_t'] = torch.zeros(T)
        for t in reversed(range(T)):
            R = H['r'][t] + gamma * R
            H['R_t'][t] = R
            

        val_avg_loss = 0.0        
        states_tensor = torch.as_tensor(np.asarray(H['s']), dtype=torch.float32)
        
        for epoch in range(val_fn_epochs):
            H['V_pred'] = value_fn(states_tensor).view(-1)
            
            value_optimizer.zero_grad()
            
            val_loss = mse_loss(H['V_pred'], H['R_t']) 
            val_avg_loss += val_loss.item()
            
            val_loss.backward()
            value_optimizer.step()
            
        val_avg_loss /= val_fn_epochs
        stats['val_loss'] += [val_avg_loss]

        policy_optimizer.zero_grad()

        loss = 0
        for t in range(T):
            advantage = H['R_t'][t] - value_fn(states_tensor[t]).detach()
            loss -= H['log_probs'][t] * advantage

        loss.backward()
        policy_optimizer.step()

    return stats
```

## Linear Policy 

```python
s_dims = env.observation_space.shape[0]
a_dims = env.action_space.n

lin_policy = LinearPolicy(s_dims, a_dims)

stats_lin = VanillaPolicyGradient(env, lin_policy, **CONFIG['VPG']['linear'])
```

```python
animate_policy(env, lin_policy, **CONFIG['animation'])
```

```python
plot_stats(stats_lin, CONFIG['plot_stats_lin_size'])
```

## MLP Policy

```python
s_dims = env.observation_space.shape[0]
a_dims = env.action_space.n

mlp_policy = MLPPolicy(s_dims, a_dims)
stats_mlp = VanillaPolicyGradient(env, mlp_policy, **CONFIG['VPG']['mlp'])
```

```python
animate_policy(env, mlp_policy, **CONFIG['animation'])
```

```python
plot_stats(stats_mlp, CONFIG['plot_stats_mlp_size'])
```
